```python
N = 10000
p = np.pi

theta = np.linspace(0,2*p,N)          # polar coordinates
r = 1 + (3*np.sin(3*theta))/4

x = r*np.cos(theta)                 # cartesian coordinates
y = r*np.sin(theta)

plt.plot(x,y)
```

```python
A = (1/2) * np.sum(r**2) * (theta[1]-theta[0])
A
```

```python
L = np.sum(np.sqrt(r**2 + (np.gradient(r,theta))**2)) * (theta[1]-theta[0])
L
```

###### 1. Fibonacci Series using Binet Formula

```python
import numpy as np
```

```python
n = int(input())

arr = np.arange(1,n+1)

x = np.sqrt(5)

alpha = (1 + x)/2
beta = (1 - x)/2

#calculate fibonacci numbers using numpy array

fn = (alpha**arr - beta**arr)/ x

np.rint(fn)
```
