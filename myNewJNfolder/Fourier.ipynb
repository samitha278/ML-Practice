```python
import scipy.integrate as integrate 
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-20,20,100000)

w = np.pi/3

ft = 2 + 2*np.sin(w*t) + 5*np.cos(w*t) + 3*np.sin(2*w*t)+ 6*np.cos(2*w*t) + 3*np.sin(3*w*t) + 7*np.cos(3*w*t)


plt.plot(t, ft )
plt.grid()

plt.show()
```

```python
T = (2*np.pi) / w
f = 1/T
```

## Fourier Series

```python
dx = 40 / 100000

a_0 = (np.sum([ft[50000: int(50000 + 2500*T)]]) * dx)/T
a_0
```

```python
cos = []
sin = []


for i in range(1,10):
    
    #cos
    cos_a = np.cos(w*i*t)
    
    ft_cost = ft * cos_a
    
    a_i = 2*(np.sum([ft_cost[50000: int(50000 + 2500*T)]]) * dx)/T
    
    #sin
    sin_a = np.sin(w*i*t)
    
    ft_sint = ft * sin_a
    
    b_i = 2*(np.sum([ft_sint[50000: int(50000 + 2500*T)]]) * dx)/T 
    
    cos.append(round(a_i,3))
    sin.append(round(b_i,3))
    
    
sin,cos
```
