# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1,20,100)
y = np.sin(x) + 0.1 * np.random.randn(len(x))
```

```python
y2 = np.sin(x)
```

```python
plt.figure(figsize = (10,5))

plt.plot(x,y,'o',color = "purple" , lw = 0.7 , ms = 3 , label = "component 1")

plt.plot(x,y2,'-',color = "red" , lw = 0.7 , ms = 3 , label = "component 2")

plt.xlabel("x" , fontsize = 10)
plt.ylabel("y",fontsize = 10)

plt.ylim(top = 1.5)

plt.legend(loc = "upper right" , fontsize = 10 , ncol = 2)
```

## Advanced API

```python
fig , axes = plt.subplots(2,2, figsize = (10,5))


x = np.linspace(0,10,100)
y = np.random.randn(100) + 0.5

a1 = axes[0][1]
a1.plot(x,y)
a1.tick_params(axis = "both" , labelsize = 5)
a1.set_title("Normal plot")

a2 = axes[1][0]
a2.hist(y, bins = 100 , density = True , histtype = "step")
a2.set_title("Histogram")

plt.show()
```

## 2D plots

```python
_ = np.linspace(-5,5 , 100)

x,y = np.meshgrid(_,_)
z = x**2 + x*y

plt.contourf(x,y,z , levels = 30)
plt.colorbar(label = "volatge [V]")
plt.show()
```

```python
cs = plt.contour(x,y,z , levels = 20)
plt.clabel(cs, fontsize = 8)
plt.show()
```

## 3D surface plot

```python
fig , ax = plt.subplots(subplot_kw = {"projection": "3d"})

ax.plot_surface(x,y,z , cmap = "coolwarm")
plt.show()
```

## Stream plot 
###### vector field

```python
_ = np.linspace(-3,3,100)

x,y = np.meshgrid(_,_)
u = x**2 + y
v = x - y**2

speed = np.sqrt(u**2 + v**2)
```

```python
fig , axes = plt.subplots(2,2, figsize = (7,7))

a1 = axes[0][0]
a1.streamplot(x,y,u,v)

a2 = axes[0][1]
a2.streamplot(x,y,u,v , color = speed)

a3 = axes[1][0]
a3.streamplot(x,y,u,v , color = speed , linewidth = 5 * speed / speed.max())

a4 = axes[1][1]
a4.streamplot(x,y,u,v , color = speed , start_points = np.array([[1,0],[0,1]]))

plt.show()
```
